```python
import cv2
import numpy as np
cap = cv2.VideoCapture(0)

def nothing(x):
    pass
# Creating a window for later use
cv2.namedWindow('Result')
cv2.namedWindow('Parameter1')
cv2.namedWindow('Parameter2')
cv2.namedWindow('Candidate')
cv2.namedWindow('Target')
# Starting with 100's to prevent error while masking
h_Max,h_Min,s_Max,s_Min,v_Max,v_Min = 179,0,255,0,255,0
# Creating track bar

#For contour finding
cv2.createTrackbar('h_Max', 'Parameter1',0,179,nothing)
cv2.createTrackbar('h_Min', 'Parameter1',0,179,nothing)
cv2.createTrackbar('s_Max', 'Parameter1',0,255,nothing)
cv2.createTrackbar('s_Min', 'Parameter1',0,255,nothing)
cv2.createTrackbar('v_Max', 'Parameter1',0,255,nothing)
cv2.createTrackbar('v_Min', 'Parameter1',0,255,nothing)
cv2.createTrackbar('Threshold', 'Parameter1',0,1,nothing)
cv2.createTrackbar('Blur', 'Parameter1',0,1,nothing)
cv2.createTrackbar('Dilate', 'Parameter1',0,1,nothing)
cv2.createTrackbar('Open', 'Parameter1',0,1,nothing)
cv2.createTrackbar('Close', 'Parameter1',0,1,nothing)
cv2.createTrackbar('Erode', 'Parameter1',0,1,nothing)
cv2.createTrackbar('FindCountours', 'Parameter1',0,1,nothing)
cv2.createTrackbar('Area_Low', 'Parameter1',0,10000,nothing)

#For cropped Image
cv2.createTrackbar('h_Max', 'Parameter2',0,179,nothing)
cv2.createTrackbar('h_Min', 'Parameter2',0,179,nothing)
cv2.createTrackbar('s_Max', 'Parameter2',0,255,nothing)
cv2.createTrackbar('s_Min', 'Parameter2',0,255,nothing)
cv2.createTrackbar('v_Max', 'Parameter2',0,255,nothing)
cv2.createTrackbar('v_Min', 'Parameter2',0,255,nothing)
cv2.createTrackbar('Threshold', 'Parameter2',0,1,nothing)
cv2.createTrackbar('Blur', 'Parameter2',0,1,nothing)
cv2.createTrackbar('Dilate', 'Parameter2',0,1,nothing)
cv2.createTrackbar('Open', 'Parameter2',0,1,nothing)
cv2.createTrackbar('Close', 'Parameter2',0,1,nothing)
cv2.createTrackbar('Erode', 'Parameter2',0,1,nothing)
cv2.createTrackbar('FindCountours', 'Parameter2',0,1,nothing)
cv2.createTrackbar('Area_Low', 'Parameter2',0,10000,nothing)
cv2.createTrackbar('Number', 'Parameter2',0,9,nothing)

#Neural Network Parameter : Training Accuracy:1 Testing Accuracy: 0.8
W = np.load('W.npy')
b = np.load('b.npy') 
dim = (28,28)
class Filter:
    def __init__(self,img):
        self.img = img
        self.h_Max = 179
        self.s_Max = 255
        self.v_Max = 255
        self.h_Min = 0
        self.s_Min = 0
        self.v_Min = 0
        self.Threshold = 1
        self.Blur = 1
        self.Dilate = 1
        self.Open = 1
        self.Close = 1
        self.Erode = 1
        self.FindCountours = 1
        self.Area_Low = 0
    def fixparameter(self,h_Max,h_Min,s_Max,s_Min,v_Max,v_Min,Threshold,Blur,Dilate,Open,Close,Erode,FindCountours,Area_Low):
        self.h_Max,self.h_Min,self.s_Max,self.s_Min = h_Max,h_Min,s_Max,s_Min
        self.v_Max,self.v_Min,self.Threshold,self.Blur,self.Dilate = v_Max,v_Min,Threshold,Blur,Dilate
        self.Open,self.Close,self.Erode,self.FindCountours,self.Area_Low = Open,Close,Erode,FindCountours,Area_Low
    def getTrackbarPos(self, switch):
        Source = 'Parameter'+str(switch)
        self.h_Max = cv2.getTrackbarPos('h_Max',Source)
        self.s_Max = cv2.getTrackbarPos('s_Max',Source)
        self.v_Max = cv2.getTrackbarPos('v_Max',Source)
        self.h_Min = cv2.getTrackbarPos('h_Min',Source)
        self.s_Min = cv2.getTrackbarPos('s_Min',Source)
        self.v_Min = cv2.getTrackbarPos('v_Min',Source)
        self.Threshold = cv2.getTrackbarPos('Threshold',Source)
        self.Blur = cv2.getTrackbarPos('Blur',Source)
        self.Dilate = cv2.getTrackbarPos('Dilate',Source)
        self.Open = cv2.getTrackbarPos('Open',Source)
        self.Close = cv2.getTrackbarPos('Close',Source)
        self.Erode = cv2.getTrackbarPos('Erode',Source)
        self.FindCountours = cv2.getTrackbarPos('FindCountours',Source)
        self.Area_Low = cv2.getTrackbarPos('Area_Low',Source)
    def Filter(self):
        im = cv2.bitwise_and(self.img,self.img,mask = self.GetMask()) #HSL Filtering
        im = self.BLUR(im) #Blur image
        im = self.THRESHOLD(im) #Adaptive Threshold
        #Morphological Transformations
        im = self.CLOSE(im)
        im = self.DILATE(im)
        im = self.ERODE(im)
        im = self.OPEN(im)
        self.img = im
        return im
    def GetMask(self):
        hsv = cv2.cvtColor(self.img,cv2.COLOR_BGR2HSV)
        lower_blue,upper_blue = self.bound()
        mask = cv2.inRange(hsv,lower_blue, upper_blue)
        return mask
    def bound(self):
        lower_blue = np.array([self.h_Min,self.s_Min,self.v_Min])
        upper_blue = np.array([self.h_Max,self.s_Max,self.v_Max])
        return lower_blue,upper_blue
    # get info from track bar and appy to result

    def THRESHOLD(self,img):
        if self.Threshold == 1: #Switch
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
            Tmask = cv2.adaptiveThreshold(gray,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,\
            cv2.THRESH_BINARY,11,2)
            imgT = cv2.bitwise_and(img,img,mask = Tmask)
            return imgT
        else:
            return img
    def BLUR(self,img):
        if self.Blur == 1: #Switch
            imgB = cv2.medianBlur(img,5)
            return imgB
        else:
            return img
    def DILATE(self,img):
        if self.Dilate == 1:#Switch
            kernel = np.ones((10,10),np.uint8)
            imgDilate = cv2.dilate(img,kernel,iterations = 1)
            return imgDilate
        else:
            return img
    def OPEN(self,img):
        if self.Open == 1:#Switch
            kernel = np.ones((10,10),np.uint8)
            imgO = cv2.morphologyEx(img, cv2.MORPH_OPEN, kernel)
            return imgO
        else:
            return img
    def ERODE(self,img):
        if self.Erode == 1:#Switch
            kernel = np.ones((5,5),np.uint8)
            imgE = cv2.erode(img,kernel,iterations = 1)
            return imgE
        else:
            return img
    def CLOSE(self,img):
        if self.Close == 1:#Switch
            kernel = np.ones((10,10),np.uint8)
            imgC = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel)
            return imgC
        else:
            return img

    def countour(self):
        gray = cv2.cvtColor(self.img, cv2.COLOR_BGR2GRAY)
        _,contours,_ = cv2.findContours(gray,cv2.RETR_LIST,cv2.CHAIN_APPROX_SIMPLE)
        return contours

    def neuralprocess(self,W,b,dim):
        im = cv2.cvtColor(self.img, cv2.COLOR_BGR2GRAY)
        Input = cv2.resize(im,dim, interpolation = cv2.INTER_AREA)
        Input = Input.reshape((1,dim[0]*dim[1]))
        output = np.maximum((np.matmul(Input, W)+b),0)
        return np.argmax(output)
    
    
while(1):
    _, frame = cap.read()
    # Normal masking algorithm
    IMAGE1 = Filter(frame)
    IMAGE1.getTrackbarPos(1)
    #IMAGE1.fixparameter(h_Max=150,h_Min=116,s_Max=190,s_Min=54,v_Max=255,v_Min=108,Threshold=0,Blur=1,Dilate=1,Open=0,Close=0,Erode=0,FindCountours=1,Area_Low=1000)
    IMAGE1.Filter()
    #Label Contours
    if IMAGE1.FindCountours==1:
        for cnt in IMAGE1.countour(): 
            x,y,w,h = cv2.boundingRect(cnt)
            IMAGE2 = Filter(frame[y:y+h,x:x+w])
            IMAGE2.getTrackbarPos(2)
            #IMAGE2.fixparameter(h_Max=179,h_Min=115,s_Max=255,s_Min=0,v_Max=255,v_Min=0,Threshold=0,Blur=0,Dilate=0,Open=0,Close=0,Erode=0,FindCountours=0,Area_Low=0)
            IMAGE2.Filter()
            kernel = np.ones((5,5),np.uint8)
            Cand = cv2.cvtColor(IMAGE2.img, cv2.COLOR_BGR2GRAY)
            if cv2.contourArea(cnt)>=IMAGE2.Area_Low:
                if IMAGE2.neuralprocess(W,b,dim) == cv2.getTrackbarPos('Number','Parameter2'): #Represents yes
                    cv2.imshow('Target',Cand)
                    cv2.rectangle(IMAGE1.img,(x,y),(x+w,y+h),(255,255,255),2)
                else:
                    cv2.rectangle(IMAGE1.img,(x,y),(x+w,y+h),(0,0,255),2)
                    cv2.imshow('Candidate',Cand)
    cv2.imshow('Parameter1',IMAGE1.img)
    cv2.imshow('Result',IMAGE1.img)
    cv2.imshow('Original',frame)
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break

cap.release()
cv2.destroyAllWindows()
del IMAGE1,IMAGE2
```
